```python
# import all the dataset

import pandas as pd

hour_data=pd.read_csv("hourly_data.csv",low_memory=False)

day_data=pd.read_csv("daily_data.csv")

month_data=pd.read_csv("monthly_data.csv")

three_hour_data=pd.read_csv("three_hour_data.csv")
```

```python
#cleaning the dataset

day_data['DailyPrecipitation']=day_data['DailyPrecipitation'].replace('T','0.01')
day_data['DailyPrecipitation']=pd.to_numeric(day_data['DailyPrecipitation'])

day_data['DailySnowfall']=day_data['DailySnowfall'].replace('T','.01')
day_data['DailySnowfall']=pd.to_numeric(day_data['DailySnowfall'])

# all dataset into singleset

list_of_details=[hour_data,day_data,month_data,three_hour_data]
```

```python
#iterator all list using for loop

for i, df in enumerate(list_of_details):

        num_cols=df.select_dtypes(include='number')

        df=df.drop_duplicates()

        describe_data =df.describe()

        mean_val = num_cols.mean().astype(int)

        med_val = num_cols.median(numeric_only = True).astype(int)

        std_val = num_cols.std(numeric_only = True)

        mode_val = num_cols.mode()

        #EDR Basic operation
        print("\nMEAN")
        print(mean_val)
        print(" ")
        print("\nMEDIAN")
        print(med_val)
        print(" ")
        print("\nSTANDARD DEVIATION")
        print(std_val)
        print(" ")
        print("\nMODE")
        print(mode_val)
```

```python

for i, df in enumerate(list_of_details):

        num_cols=df.select_dtypes(include='number')

        df['DATE']=pd.to_datetime(df['DATE'])

        df=df.drop_duplicates()

        range_val = num_cols.max(numeric_only = True) - num_cols.min(numeric_only = True)

        variance_val =num_cols.var(numeric_only = True)

        cor_mat = num_cols.corr(numeric_only = True)

        Q1 = num_cols.quantile(0.25,numeric_only = True)
        Q3 = num_cols.quantile(0.75,numeric_only = True)
        iqr = Q3 - Q1
        #Basic EDR operation
        print("\nRANGE")
        print(range_val)
        print(" ")
        print("\nVARIANCE")
        print(variance_val)
        print(" ")
        print("\nCORRELATION MATRIX:")
        print(cor_mat)
        print(" ")
        print("\nIQR")
        print(iqr)
```
